## Data Cleaning

```python
# drop blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# quick stats for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
# drop duplicates
df.drop_duplicates()
```

# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and their count of orders

ax = sns.countplot(x = 'Gender',data = df, width = 0.2, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

*Females are in comparison to males more likely to place orders*

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, width = 0.2, hue = 'Gender')
```

*From above graphs it is obvious that the females spent more amount than males*

### Age

```python
# plotting a bar chart for Gender and their count of orders distributed in different age groups
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', order = ['0-17','18-25','26-35','36-45','46-50','55+'])

for bars in ax.containers:
    ax.bar_label(bars)
```

*The age group which orders most is aged between 26 and 35 and as established before, females order more than males in every age group*

```python
df.groupby(['Age Group'])['Amount'].sum().reset_index()
```

```python
# plotting a bar chart for Gender and amount spent distributed in different age groups
sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index()

sns.set(rc={'figure.figsize':(9,5)})
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group', width = 0.9)
```

*From above graphs we can see that the customers of age group between 26-35 yrs spends the highest amount/money*

### State

```python
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'State',y= 'Orders', hue = 'State', width = 0.9)
```

*Highest number of orders statewise*

```python
# total amount/sales from top 5 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue = 'State', width = 0.5)
```

*From above graphs we can see that most sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

*Married people tend to order more than unmarried people*

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married ones and that too females and they have higher purchasing power*

### Occupation

```python
# number of order by occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

*From the above figure it can be deduced that most number of orders are from employees of IT Sector, Aviation, Healthcare, Banking & Govt*

```python
# total amount by occupation
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue = 'Occupation', width = 0.7)
```

*From above graph we can see that the highest amounts are spent by employees of IT, Healthcare and Aviation sector*

### Product Category

```python
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category', width = 0.9)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', hue = 'Product_Category')
```

*From above graphs we can see that most of the sold products by order and amount are from Food, Clothing and Electronics category*

### Most sold products

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue = 'Product_ID')
```

*These are the Product IDs of the most sold products by their quantity and there is the pattern that the highest orders tend to have highest amount/revenue*
